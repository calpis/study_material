```python
import numpy as np
import csv
train_data = list()
import io

with io.open('/Users/momori/documents/github_mcafee/mobile_cloud/APKAnalyzer/signature_match_script/resources/default_apks/r.txt','r', encoding="utf-8") as f:
    common_text = f.read().split()
common_text = set(common_text)
#str.decode('cp1252').encode('utf-8')
with io.open(train_data_file, 'r', encoding='utf-8') as f:
    train_data = f.read().split('\n')
#train_data = [unicode(i, errors='ignore') for i in train_data]

try:
    train_data = [line.encode('utf-8') for line in train_data]
    train_data = [unicode(line) for line in train_data]
except:
    pass

# print train_data
# print len(train_data)
train_str, train_count = np.unique(train_data, return_counts=True)
train_str_count = zip(train_str, train_count)

string_count_dict = dict()
for k,v in train_str_count:
    string_count_dict[k] = v

train_data = list(set([i.replace('\r','') for i in train_data]))

score_test_candidate = []
def similarity(v1, v2):
    n1 = np.linalg.norm(v1)
    n2 = np.linalg.norm(v2)
    res = np.dot(v1, v2)/n1/n2
    return res

for test_word in test_words:
    if 'abc' in test_word:
        continue
    if test_word in common_text:
        continue
    max_sim = 0
    best_word = ''
    
    v2 = model.get_word_vector(test_word)
    for i in train_data:
        v1 = model.get_word_vector(i)

        sim = similarity(v1, v2)
        if sim>max_sim:
            max_sim=sim
            best_word = i
    score_test_candidate.append((max_sim, test_word, best_word, string_count_dict[best_word]))
```
