## MLE (maximum likelihood estimator) example

Consider a regular coin with two possible outcomes, heads (H) or tails (T). Through 100 flips, we observed 55 heads. We want to find P(H) which maximizes the probability of the observed data.

Given the prob_bernoulli function, we want to find the derivative, set it to 0 to find the p that maximizes the data observed



```python
import numpy as np
import math

from scipy.misc import derivative
import  matplotlib.pyplot as plt
```

```python
num_trials = 100
num_heads = 55


def nCr(n,r):
    f = math.factorial
    return f(n) / f(r) / f(n-r)

def prob_bernoulli(prob):
    return nCr(num_trials, num_heads)*math.pow(prob,num_heads)*math.pow((1-prob),(num_trials-num_heads))
```

```python
(prob_bernoulli(0.55))
```

```python
probabilities = np.linspace(0, 1, 101)
prob_d_given_p = [prob_bernoulli(i) for i in probabilities]
```

```python
plt.plot(probabilities, prob_d_given_p)
plt.show()
```

## log-likelihood

since log is an increasing function, the maxima of the likelihood and log likelihood are the same

```python
def log_likelihood(prob):
    if prob==0.0:
        prob = 0.0001
    if prob==1:
        prob = 0.999
    return math.log(prob_bernoulli(prob),2)

prob_d_given_p = [log_likelihood(i) for i in probabilities]
plt.plot(probabilities, prob_d_given_p)
plt.show()
```
